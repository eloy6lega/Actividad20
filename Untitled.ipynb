# Leer CSVs y representar Gr√°ficos

```python
import pandas as pd
import matplotlib.pyplot as plt

def parte1():
    datos=pd.read_csv('products.csv')
    #print(datos)
    df=datos[['unitsInStock','categoryID']]
    #print(df)
    df.head(25).plot.barh(x="unitsInStock", y="categoryID")
    plt.show()
parte1()
```

```python
def parte2():
    datos=pd.read_csv('products.csv')
    #print(datos)
    df= datos[['unitPrice','categoryID']]
    #print(df)
    df.unitPrice.head(30).value_counts().plot.pie()
    plt.show()
parte2()
```

```python
def parte3():
    datos=pd.read_csv('products.csv')
    #print(datos)
    df= datos[['unitPrice','productName']]
    #print(df)
    df.head(30).unitPrice.plot.hist()
    plt.show()
parte3()
```
